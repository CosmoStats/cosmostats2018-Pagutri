_Reproduce las gráficas de la distribución uniforme de puntos dentro de un círculo y una esfera._

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
import pylab
import math
import random

%matplotlib inline
pylab.rcParams['figure.figsize'] = (10, 6)
```

```python
rand_x = []
rand_y = []
uniform_x = []
uniform_y = []
lenght = 10000
radius = 1.0

for i in range(lenght):
    rand_x.append(2.0 * radius * random.random() - radius)
    rand_y.append(2.0 * radius * random.random() - radius)
    
for i in range(lenght):
    x = 2.0 * radius * random.random() - radius
    y = 2.0 * radius * random.random() - radius
    
    if math.sqrt(x**2 + y**2) < radius**2:
        uniform_x.append(x)
        uniform_y.append(y)
        
plt.plot(rand_x, rand_y, 'r,')
plt.plot(uniform_x, uniform_y, 'b,')
plt.gca().set_aspect('equal')
plt.xlabel("x")
plt.ylabel("y")
```

```python
X = []
Y = []
Z = []
lenght = 1000
radius = 1.0

for i in range(lenght):
    x = 2.0 * radius * random.random() - radius
    y = 2.0 * radius * random.random() - radius
    z = 2.0 * radius * random.random() - radius
    
    if math.sqrt(x**2 + y**2 + z**2) < radius**2:
        X.append(x)
        Y.append(y)
        Z.append(z)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X, Y, Z, c = 'b', marker = '.') 
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.gca().set_aspect('equal')
```
