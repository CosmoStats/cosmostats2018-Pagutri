# Histogramas de distribuciones de puntos aleatorios

#### 3. Construye una caja de lado 1000 en 2 y 3 dimensiones de una distribución aletatoria de puntos.

Usaré una distribución aleatoria uniforme.

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams
import math
from random import random

rcParams['figure.figsize'] = (10, 6)
```

En dos dimensiones:

```python
rand_x = []
rand_y = []
lado = 1000
n_puntos_2d = 2000

for i in range(n_puntos_2d):
    rand_x.append(lado * (random() - 0.5))
    rand_y.append(lado * (random() - 0.5))
        
plt.scatter(rand_x, rand_y, s = 1)
plt.gca().set_aspect('equal')
plt.xlabel("x")
plt.ylabel("y")
```

En tres dimensiones:

```python
X = []
Y = []
Z = []
lado = 1000
n_puntos_3d = 2000

for i in range(n_puntos_3d):
    X.append(lado * (random() - 0.5))
    Y.append(lado * (random() - 0.5))
    Z.append(lado * (random() - 0.5))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X, Y, Z, c = 'b', marker = '.') 
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.gca().set_aspect('equal')
```

#### 4. Código que calcule histograma de distancias para 2d y 3d de una distribución de puntos aleatorios, escogiendo el tamaño de los bins, y normalizada a que su área sea 1.

Utilizaré las distribuciones generadas arriba.

En dos dimensiones:

```python
dist_max = math.sqrt(2.0 * lado**2)
distancias_2d = []
tamanio_bin = 50

for i in range(n_puntos_2d - 1):
    for j in range(i + 1, n_puntos_2d):
        distancia = math.sqrt((rand_x[i] - rand_x[j])**2 + (rand_y[i] - rand_y[j])**2)
        distancias_2d.append(distancia)

bins = np.arange(0.0, dist_max, tamanio_bin);
plt.hist(distancias_2d, bins=bins, density = True, alpha=0.5) # Density normaliza por mí
```

En tres dimensiones:

```python
dist_max = math.sqrt(3.0 * lado**2)
distancias_3d = []
tamanio_bin = 50

for i in range(n_puntos_3d - 1):
    for j in range(i + 1, n_puntos_3d):
        distancia = math.sqrt((X[i] - X[j])**2 + (Y[i] - Y[j])**2 + (Z[i] - Z[j])**2)
        distancias_3d.append(distancia)

bins = np.arange(0.0, dist_max, tamanio_bin);
plt.hist(distancias_3d, bins=bins, density = True, alpha=0.5)
```
